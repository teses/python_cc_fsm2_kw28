Memberchip Operator

Der Operator prüft ob ein Wert in einer Liste oder Dictionary vorhanden sind

```python
meineListe = ["Apfel", "Banane", "Kiwi"]
if "Banane" in meineListe:
    print("Banane ist in der Liste")
```

```python
meineListe = ["Apfel", "Banane", "Kiwi"]
suchwort = "Apfel"
if suchwort in meineListe:
    print(suchwort + " ist in der Liste")
```

Suchen im dictionary

```python
person = {
    "firstName": "Susi",
    "lastName": "Sauer",
}

if "Susi" in person.values():
    print("Susi gefunden")
```
