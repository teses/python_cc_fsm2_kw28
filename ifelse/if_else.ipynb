IF-Else

```python
alter = 17

if alter >= 18:
    print("tanzen....")
else:
    print("ab nach hause...")
```

If-elif

```python
a = 5
b = 10

if a < b:
    print("a kleiner b")
elif a == b:
    print("a is gleich b")
else:
    print("a größer b")
```
