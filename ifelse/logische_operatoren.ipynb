logische Operatoren

or - ODER
and - UND
not - NICHT

xor GIBT es nicht

Bei komplizierten Bedingungen (groß und viel) ist ein eWahrheitstabelle nützlich.

```python
A = True
B = True
C = True

if (not A and B and not C) or (A and B and not C) or (A and not B and C)  or (A and B and C)  :
    print("X = 1")
    # 20 zeilen code
else:
    print("X = 0")
```
