Identity Operator
Der Operator vergleicht zwei Objekte/Variablen miteinander.
Es wird geprüft, ob dieser ein und dasselbe Objekt im Speicher sind (gleiche Speicherreferenz).

```python
meineListe = ["Hallo", "Welt"]
meineListe2 = meineListe

if meineListe is meineListe2:
    print("1. meineListe und meineListe2 sind die selben")


meineListe2.append("Test")
if meineListe is meineListe2:
    print("2. meineListe und meineListe2 sind die selben")


print(meineListe)
print(meineListe2)
```

Nochmal wenn die Listen die gleichen sind

```python
testlist1 = ["Hallo", "Welt"]
testlist2 = ["Hallo", "Welt"]
if testlist1 is testlist2:
    print("testlist1 und testlist2 sind gleich")
```

Achtung. Der is-Operator nur bei komplexen Datentypen

```python
a = 5
b = 5

if a is b:
    print("1. das selbe")

b += 1
if a is b:
    print("2. das selbe")
```
