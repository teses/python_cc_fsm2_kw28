In Python bezeichnet man als Callback-Funktion eine Funktion, die als Argument an eine andere Funktion übergeben wird
und dort zu einem späteren Zeitpunkt aufgerufen wird.

Man verwendet sie oft, um Verhalten flexibel zu gestalten, z.B. bei Sortierung, Event-Handling oder asynchronen Operationen.


```python

from typing import Callable

def FunktionA():
    print("Hallo ich bin in der FunktionA()")


def FunktionB():
    print("Hallo ich bin in der FunktionB()")


def Hauptfunktion(func: Callable):
    print("Hallo ich bin in der Hauptfunktion()")
    func()
    print("Die Aufgerufene Funktion func() heißt " + func.__name__)




Hauptfunktion( FunktionA )
Hauptfunktion( FunktionB )
```
