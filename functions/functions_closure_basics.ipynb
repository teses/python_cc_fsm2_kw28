Ein Closure in Python ist eine Funktion, die sich Variablen aus ihrem umgebenden Gültigkeitsbereich (Enclosing Scope) merkt,
selbst nachdem die äußere Funktion bereits beendet wurde.

Ein Closure besteht aus:
1. einer äußeren Funktion,
2. einer inneren Funktion,
3. einer oder mehreren Variablen der äußeren Funktion, die von der inneren Funktion verwendet werden,
4. der Rückgabe der inneren Funktion.

Closures eignen sich besonders für:

- Callback-Funktionen
- Factory-Funktionen
- Dekoratoren

```python
# Einfaches Beispiel
# Obwohl begruessung() bereits beendet ist, kennt hallo() den Wert von name.
def begruessung(name):

    def hallo():
        print(f"Hallo {name}!")

    return hallo

spreche = begruessung("Thomas")
spreche()

spreche2 = begruessung("Susi")
spreche2()
```

# Beispiel mit doppelten return

```python
def aussen(name):
    def innen():
        return f"Hallo aus innen mit name: {name}"
    return innen  # Die Funktion selbst zurückgeben

f = aussen("Thomas")
ret = f() # "Hallo aus innen!"
print(ret)
```

# Factory Funktionen

```python
def multiplikator(faktor):
    def multipliziere(x):
        return x * faktor
    return multipliziere

mal3 = multiplikator(3)
mal5 = multiplikator(5)
# können hunderte zeilen liegen

print(mal3(10))  # 30
print(mal5(10))  # 50
```

Factory Beispiel

```python
def mwst_rechner(satz):
    def berechne(x):
        return x * (1 + satz/100)
    return berechne


mwst19 = mwst_rechner(19)
mwst7 = mwst_rechner(7)

print(mwst19(100))
print(mwst7(100))
```
