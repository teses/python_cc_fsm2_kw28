In Python kannst du Funktionen innerhalb von Funktionen definieren.

Das nennt man verschachtelte Funktionen oder innere Funktionen

Praktische Anwendung:

- Funktionen kapseln
- Hilfsfunktionen, die nur lokal gebraucht werden
- Closures (Daten + Funktion bündeln)
- Dekoratoren (werden mit inneren Funktionen umgesetzt)

Wichtige Punkte:
Innere Funktion ist nur im Gültigkeitsbereich der äußeren Funktion verfügbar

```python
# Das Grundprinzip:
def aussen():
    print("Ich bin die äußere Funktion.")

    def innen():
        print("Ich bin die innere Funktion.")

    # Aufruf der inneren Funktion innerhalb der äußeren
    innen()


aussen()
```
