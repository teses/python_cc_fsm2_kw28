```python
def addressoutput3(zip: str, street:str, city:str) -> str:
    """
    Function to format an address string.
    Args:
        zip (str): The zip code of the address.
        street (str): The street name of the address.
        city (str): The city of the address.
    Returns:
        str: The formatted address string.
    """
    return f"{street}, {city}, {zip}"


adresse = addressoutput3(zip="01234", city="Entenhausen", street="Musterstr. 1")

print(adresse)
```
