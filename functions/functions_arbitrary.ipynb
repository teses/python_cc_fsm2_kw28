# Willkürliche Argumente in einer Funktion

Mit positionierten Argumenten

- *args - Parameter
- arbitrary arguments

```python

def kinder(*args):
    print(args)  # args ist jetzt ein tuple mit den werten
    for item in args:
        print(item)


kinder("Susi", "Hans", "Laura", "Jasmin")
# kinder("Tick", "Trick", "Track")
# kinder("Max", "Moritz")
# kinder("Donald")

```

# Willkürliche Argumente in einer Funktion

Mit benannten Argumenten

- **kwargs - Parameter
- named arguments

```python
def addressOutput(**kwargs):
    print(kwargs)

    print(kwargs['street'])
    print(kwargs['zip'] + " " + kwargs['city'])


addressOutput(street="Musterstrasse", zip="12345", city="Entenhausen")
```

Das Proplem sind hier die Default werte

```python
def addressOutput(**kwargs):
    """

    Folgende Datren werden erwartet
    {
        "street" : "",
        "zip" : "",
        "city" : "",
    }


    :param kwargs:
    :return:
    """
    if "street" not in kwargs.keys():
        kwargs['street'] = "Defaultstrasse"

    if "zip" not in kwargs.keys():
        kwargs['zip'] = "12345"

    if "city" not in kwargs.keys():
        kwargs['city'] = "City"


    print(kwargs['street'])
    print(kwargs['zip'] + " " + kwargs['city'])


addressOutput()
```

Bei folgenden musss man sich entscheiden

```python
def addressOutput(kwargs):
    print(kwargs)

    print(kwargs['street'])
    print(kwargs['zip'] + " " + kwargs['city'])


daten = {
    "street" : "Musterstrasse",
    "zip" : "12345",
    "city" : "Entenhausen",
}

#addressOutput(street=daten["strasse"], zip=daten["zip"], city=daten["city"])

addressOutput(daten)


```

# Die Mischung

*args, **kwargs

```python

def mein_funktion(hauptargument,  *args, **kwargs):

    print('hauptargument', hauptargument)
    print('tuple - positional arguments', args)
    print('dict - named arguments', kwargs)


# keyword argument follows positional argument
mein_funktion("hauptdaten", "A", "B", "C", street="Musterstrasse", zip="12345", city="Entenhausen")
```
