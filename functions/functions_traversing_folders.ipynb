```python
import os

def zeige_verzeichnis(pfad, indent=0):
    """Zeigt Verzeichnisstruktur rekursiv an"""
    try:
        for element in os.listdir(pfad):
            element_pfad = os.path.join(pfad, element)
            print(" " * indent + element)
            if os.path.isdir(element_pfad):
                zeige_verzeichnis(element_pfad, indent + 1)

    except PermissionError:
        print(" " * indent + "[Keine Berechtigung]")



zeige_verzeichnis("./", 4)

```

# Rekursive Dateisuche mit callback

```python
import os

def jpg_searcher(filename, filename_pfad):
    if filename[-4:] == ".jpg":
        return True
    else:
        return False


def jupyther_searcher(filename, filename_pfad):
    if filename[-6:] == ".ipynb":
        return True
    else:
        return False


def zeige_verzeichnis(pfad, func, indent=0):
    """Zeigt Verzeichnisstruktur rekursiv an"""
    try:
        for element in os.listdir(pfad):
            element_pfad = os.path.join(pfad, element)

            if func(element, element_pfad):
                print(" " * indent + element)

            if os.path.isdir(element_pfad):
                zeige_verzeichnis(element_pfad, indent + 1)

    except PermissionError:
        print(" " * indent + "[Keine Berechtigung]")



zeige_verzeichnis("./", jupyther_searcher, 4)
```
