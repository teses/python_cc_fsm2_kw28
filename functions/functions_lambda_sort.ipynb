```python

# wichtig, es wird case sensitiv sortiert - aphanumerisch
words = ["Hallo", "schöne", "Welt", "abend", "10", "1", "2"]
words.sort(reverse=True)
print(words)
```

```python
# Beispiel mit normaler Funktion
Obstkorb = ["Banane", "Zitrone", "Birne", "Kiwi", "Banane", "AI"]

def obst_sorter(obst):
    return len(obst)

Obstkorb.sort(key=obst_sorter)
print(Obstkorb)
```

```python
# mit lambda
Obstkorb = ["Banane", "Zitrone", "Birne", "Kiwi", "Banane", "AI"]
Obstkorb.sort(key=lambda obst : len(obst), reverse=True)
print(Obstkorb)
```

```python
# dictionary sortieren
personen = [
    {"name": "Anna", "alter": 28},
    {"name": "Ben", "alter": 22},
    {"name": "Clara", "alter": 35}
]
personen.sort(key=lambda p : p["alter"])
print(personen)
```
