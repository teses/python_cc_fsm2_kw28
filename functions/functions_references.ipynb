# Höhere Datentypen werden als Referenz übergeben

```python

def meine_function(liste: list):
    liste.append("Ananas")
    print("in funktion", liste)


meineListe1 = ["Apfel", "Banane", "Zitrone", "Birne", "Kiwi"]

meine_function(meineListe1)

print("nach funktion",meineListe1)

```

Was tun wenn ich die liste nicht rückwärtz (als referenz) haben möchte.

Lösung list.copy()

```python
def der_manipulator(liste: list):

    neueListe = liste.copy()  # damit man eine neue Liste bekommt muss diese kopiert werden

    neueListe.append("Ananas")
    print("in funktion", neueListe)
    return neueListe


meineListe1 = ["Apfel", "Banane", "Zitrone", "Birne", "Kiwi"]
meineListe1 = der_manipulator(meineListe1)

print("nach funktion",meineListe1)
```

# Problem bei verschachtelten listen

```python
import copy

def matrix_manipulator(matrix: list) -> list:

    matrix2 = copy.deepcopy(matrix) # tiefe Kopie (alles wird rekursiv kopiert)

    matrix2[0][1] = 5
    return matrix2


original = [[1, 2], [3, 4]]
ergebnis = matrix_manipulator(original)

print("original", original)
print("ergebnis", ergebnis)
```

```python
import copy as c

original = [[1, 2], [3, 4]]
flach = copy.copy(original)
# flach = original.copy()   # mit builtin
tief = copy.deepcopy(original)

original[0].append(99)

print(original)
print(flach)  # innere Liste wurde mitverändert
print(tief)   # vollständig unabhängig
```

# 2. Beispiel

```python
import copy

# verschachtelte Liste
original = [[1, 2], [3, 4]]

# flache Kopie (nur 1. Ebene wird kopiert)
flach = original.copy()

# tiefe Kopie (alles wird rekursiv kopiert)
tief = copy.deepcopy(original)

# Änderungen testen
original[0][0] = 99

print("Original :", original)
print("Flach    :", flach)
print("Tief     :", tief)
```
