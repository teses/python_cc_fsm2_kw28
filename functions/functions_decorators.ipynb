Ein Dekorator in Python ist eine Funktion, die eine andere Funktion (oder Methode, manchmal auch eine Klasse) als Argument entgegennimmt,
deren Verhalten erweitert oder verändert, und dann wieder zurückgibt – ohne den ursprünglichen Code der Funktion dauerhaft zu ändern.

Das macht Dekoratoren sehr nützlich für
- Logging
- Zugriffskontrolle
- Caching (Memoization)
- Messung von Laufzeiten
- Validierungen
- Code-Wiederverwendung
- ...

Grundprinzip: Ein Dekorator ist einfach eine Funktion, die eine andere Funktion "umwickelt":

Syntax: Verwendung mit der @-Schreibweise:

```python
# Eine Einfache decoration Funktion. Diese Funktion wickelt eine andere Funktion ein.
def mein_dekorator(funktion):

    def wrapper():
        print("Vor der Funktion")
        funktion()
        print("Nach der Funktion")

    return wrapper


@mein_dekorator
def sag_hallo():
    print("Hallo!")

sag_hallo()
```

```python
# So würde es aussehen mit einem normalen Closure
def mein_dekorator(funktion):

    def wrapper():
        print("Vor der Funktion")
        funktion()
        print("Nach der Funktion")

    return wrapper



def sag_hallo():
    print("Hallo!")

mein_dekorator(sag_hallo)()
```

# Decorator mit return

```python
def simple_decorator_return(func):

    def wrapper():
        print("Vor der Funktion")
        ret = func()
        print("Nach der Funktion")
        return ret

    return wrapper

@simple_decorator_return
def say_hello():
    print("say_hello aufgerufen")
    return "Hallo"

print(say_hello())
```

# Decorator mit Argumenten

```python
def simple_decorator_return(func):

    def wrapper(*args, **kwargs):
        print(f"Vor der Funktion {func.__name__} | {args} | {kwargs}")
        ret = func(*args, **kwargs)
        print(f"Nach der Funktion {func.__name__} | {ret}")
        return ret

    return wrapper


@simple_decorator_return
def say_hello(name):
    print("say_hello aufgerufen")
    return f"Hallo {name}"

print(say_hello("Thomas"))
```
