- mutable (list, dict, set)
- immutable (bool, int, float, str, tuple, frozenset,..)

jetzt will man eine eigene lokale variable in dem closure manipulieren

das geht so nicht -> Lösung nonlocal oder mutable datentyp

```python
# Fehlerhaft nonlocal fehlt
def counter():
    counter = 0

    def incr():
        # nonlocal counter
        counter += 1
        return counter

    return incr


c = counter()
print(c())
print(c())
```

```python
# mit nonlocal funktioniert es!!!
def counter():
    counter = 0

    def incr():
        nonlocal counter
        counter += 1
        return counter

    return incr


c = counter()
print(c())
print(c())
```

```python
# mit mutable datatype
# ohne nonlocal muss es ein veränderbarer Datentyp sein (mutable)
def counter():
    counter = [0] # hier !!!

    def incr():

        counter[0] += 1
        return counter

    return incr


c = counter()
print(c())
print(c())
```
