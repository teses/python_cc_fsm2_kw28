```python
# wrapper funktion
def taschenrechner(callback, zahl1, zahl2):
    return callback(zahl1, zahl2)

def addiere(x, y):
    return x + y

def subtrahiere(x , y):
    return x - y

def multipliziere(x , y):
    return x * y

def dividiere(x , y):
    return x / y



print(taschenrechner(addiere, 5, 3))
print(taschenrechner(subtrahiere, 5, 3))
print(taschenrechner(multipliziere, 5, 3))
print(taschenrechner(dividiere, 5, 3))
```
