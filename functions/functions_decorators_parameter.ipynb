# Decoratoren mit eigenen Parametern

```python
def my_decorator_factory(nachricht: str):
    print(f"in der decorator hülle {nachricht}")

    # Der tatsächliche Decorator
    def the_decorator(func):

        # Die Wrapper-Funktion
        def wrapper(*args, **kwargs):
            ergebnis = func(*args, **kwargs)
            return ergebnis

        return wrapper

    return the_decorator


@my_decorator_factory("Hallo Welt!!")
def meine_funktion(name):
    print(f"Der Name ist: {name}")

# Aufruf der dekorierten Funktion
meine_funktion("Max")


```
