# Was ist Rekursion?

Rekursion bedeutet, dass eine Funktion sich selbst aufruft, um ein Problem zu lösen.
Jede rekursive Funktion braucht mindestens zwei Dinge:
- Basisfall/Abbruchbedingung – das ist die Bedingung, bei der die Rekursion endet.
- Rekursiver Fall – hier ruft sich die Funktion selbst auf, aber mit einem kleineren/simpleren Problem.

```python
def runterzaehlen(n):
    if n == 0:          # Basisfall: wenn n 0 erreicht, stoppt die Rekursion
        print("Fertig!")
    else:
        print(n)
        runterzaehlen(n-1)  # Rekursiver Aufruf mit n-1

runterzaehlen(5)
```

```python
def summe(n):
    if n == 0:          # Basisfall: wenn n 0 ist, ist die Summe 0
        return 0
    else:
        return n + summe(n-1)  # Rekursiver Aufruf mit n-1

print(summe(5))  # Ausgabe: 15
```

Doppelte rekursion

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

fibonacci(10)
```
