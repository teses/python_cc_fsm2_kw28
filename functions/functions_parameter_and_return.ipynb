# Positionierte Parameter

Functions mit angaben des Datentypen für Parameter und Return

Die Angabe des Datentyps hilf beim dokumentieren

```python
def addiere(a: int, b: int) -> int:
    return a + b


erg = addiere(5, 3)
print(erg)
```

# Funktionen mit Default Parametern

Default parameter mussen immer am Ende der definition stehen

```python
def addiere(a: int, b:int = 5) -> int:
    return a + b

erg = addiere(5)
print(erg)
```

# Funktionen mit benannten Parametern

```python
def addiere(a: int, b:int = 5) -> int:
    return a + b

erg = addiere(b=10, a=5)
print(erg)
```

Aber Achtung. Ein Parameter ist Pflicht, wenn er keine Defaultangabe hat.

```python
def addiere(a: int, b:int = 5) -> int:
    return a + b

erg = addiere(b=10) # der Parameter a ist Pflicht, da er keine Defaultangabe hat
print(erg)
```

```python
def addressOutput(zip, street, city):
    """

    :param zip:
    :param street:
    :param city:
    :return:
    """
    print(street)
    print(zip + " " + city)

addressOutput(street="Musterstr. 1", zip="12345", city="Entenhausen")
addressOutput(zip="12345", city="Entenhausen", street="Musterstr. 1")
```
