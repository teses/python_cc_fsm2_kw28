## 1. Was ist eine Variable?

Eine Variable ist ein benannter Speicherplatz für einen Wert.

```python
alter = 49
```

Dabei gilt:

- `alter` ist der **Variablenname**.
- `49` ist der **gespeicherte Wert**.
- `=` ist der **Zuweisungsoperator**.

Das Gleichheitszeichen bedeutet hier:

> Weise den Wert auf der rechten Seite der Variablen auf der linken Seite zu.


````python
alter = 49
name = "Max"

print(alter)
print(name)
````

### Variablen können neue Werte erhalten

Wird derselben Variablen erneut ein Wert zugewiesen, wird der vorherige Wert ersetzt.


````python
zahl = 10
print(zahl)

zahl = 25
print(zahl)
````

## 2. Einfache Datentypen

Wichtige einfache Datentypen sind:

- `int` - ganze Zahl
- `float` - Kommazahl
- `str` - Zeichenkette
- `bool` - Wahrheitswert `True` oder `False`
- `None` - kein Wert vorhanden

Mit `type()` kann der Datentyp eines Wertes angezeigt werden.


````python
test_number = 5
test_decimal = 3.5
test_string = "123"
test_bool = True
test_none = None

print("test_number:", test_number, type(test_number))
print("test_decimal:", test_decimal, type(test_decimal))
print("test_string:", test_string, type(test_string))
print("test_bool:", test_bool, type(test_bool))
print("test_none:", test_none, type(test_none))
````

### Wichtig: `"123"` ist keine Zahl

Die Anführungszeichen machen aus `123` einen String.

```python
123      # int
"123"    # str
```


````python
zahl = 123
text = "123"

print(zahl, type(zahl))
print(text, type(text))
````

## 3. Python ist dynamisch und stark typisiert

Die Aussage „Python ist nicht typisiert“ ist fachlich nicht richtig.

Python ist:

- **dynamisch typisiert**: Der Datentyp wird automatisch anhand des aktuellen Wertes erkannt.
- **stark typisiert**: Unpassende Datentypen können nicht ohne Weiteres miteinander verrechnet werden.

Eine Variable kann später trotzdem einen Wert eines anderen Datentyps erhalten.


````python
wert = 5
print(wert, type(wert))

wert = "Hallo"
print(wert, type(wert))
````

## 4. Typannotationen

Mit einer Typannotation wird angegeben, welcher Datentyp vorgesehen ist.

```python
alter: int = 49
```

Typannotationen:

- verbessern die Lesbarkeit,
- helfen Entwicklungsumgebungen,
- werden von Python aber nicht automatisch erzwungen.


````python
test_number: int = 5
test_decimal: float = 3.5
test_string: str = "123"
test_bool: bool = True
test_none: None = None

print(test_number)
print(test_decimal)
print(test_string)
print(test_bool)
print(test_none)
````

## 5. Benennung von Variablen

Variablennamen sollten verständlich sein.

### Erlaubt

- Buchstaben
- Zahlen
- Unterstriche

### Nicht erlaubt

- eine Zahl am Anfang
- Leerzeichen
- Sonderzeichen wie `-`, `!` oder `?`

In Python wird für Variablen normalerweise **snake_case** verwendet.


````python
menge_der_dateien = 20
du_bist_ein_wert = "Hallo"
zahl_22 = 100

print(menge_der_dateien)
print(du_bist_ein_wert)
print(zahl_22)
````

### Schreibweisen im Vergleich

```python
menge_der_dateien   # snake_case - für Python-Variablen empfohlen
mengeDerDateien     # camelCase - in anderen Sprachen häufig
MengeDerDateien     # PascalCase - in Python meist für Klassen
```

Die ursprüngliche Aussage, alle drei Schreibweisen seien für Variablen gleich gut, ist zu pauschal.

> **Merksatz:** Für Variablen und Funktionen verwendet Python normalerweise `snake_case`.


### Abkürzungen sparsam verwenden

Kurze Namen wie `i` sind bei kleinen Schleifen üblich.

Bei normalen Variablen sind verständliche Namen besser:

```python
beschreibung_des_artikels = "..."
```

statt:

```python
articleDescr = "..."
```


## 6. Python unterscheidet Groß- und Kleinschreibung

Python ist **case-sensitive**.

Diese beiden Variablen sind verschieden:


````python
VariableA = "Hallo"
variableA = "Welt"

print(VariableA)
print(variableA)
````

## 7. Reservierte Wörter

Wörter wie diese haben in Python bereits eine feste Bedeutung:

```text

```

Sie dürfen nicht als Variablennamen verwendet werden.

## 8. Konstanten als Vereinbarung

Python besitzt keine technisch unveränderbaren Konstanten.

Werte, die sich nicht ändern sollen, werden üblicherweise komplett groß geschrieben:


````python
MEHRWERTSTEUER = 0.19
VERSANDKOSTEN = 4.90

print(MEHRWERTSTEUER)
print(VERSANDKOSTEN)
````

Die Großschreibung ist eine Vereinbarung für Menschen. Python verhindert eine spätere Änderung nicht.


## 9. Eingaben mit `input()`

Mit `input()` kann der Benutzer einen Wert eingeben.

Wichtig:

> `input()` liefert immer einen String.

Auch die Eingabe `100` ist zunächst Text.


````python
eingabe = input("Bitte etwas eingeben: ")

print(eingabe)
print(type(eingabe))
````


## 10. Typumwandlung

Wichtige Funktionen:

- `int()`
- `float()`
- `str()`
- `bool()`


````python
print(int(5.5))
print(float("5.5"))
print(str(5.5))
print(int(True))
print(bool(0))
print(bool(1))
````

### Warum `10 + "Hallo"` nicht funktioniert

Das ursprüngliche Beispiel enthält absichtlich einen Fehler:

```python
10 + "Hallo"
```

- `10` ist ein `int`.
- `"Hallo"` ist ein `str`.

Die beiden Datentypen können nicht direkt addiert werden.

So funktioniert die Textausgabe:


````python
zahl = 10
text = "Hallo"

print(str(zahl) + " " + text)
````

### Ungültige Typumwandlung

Auch das funktioniert nicht:

```python
int("Hallo")
```

Der Text enthält keine gültige Zahl.

Die Behandlung solcher Fehler mit `try` und `except` folgt später.


### `int()` schneidet Nachkommastellen ab

`int()` rundet eine Kommazahl nicht normal. Der Nachkommateil wird abgeschnitten:


````python
print(int(5.9))      # 5
print(int(5.1))      # 5
print(int(-5.9))     # -5
````

Zum normalen Runden ist `round()` gedacht:

```python
print(round(5.9))  # 6
```


### Kleine Falle bei `bool()`

Leere Werte werden meistens als `False` behandelt. Nicht leere Werte gelten meistens als `True`.


````python
print(bool(""))          # False: leerer String
print(bool("False"))     # True: nicht leerer String
print(bool(0))           # False
print(bool(1))           # True
````

`"False"` ist hier nur ein Text. Weil dieser Text nicht leer ist, wird daraus `True`. Klingt erst einmal seltsam, ist aber wichtig zu wissen.


### Punkt oder Komma bei Kommazahlen?

Python erwartet einen Punkt:

```python
preis = float("9.99")
```

Eine Eingabe mit deutschem Komma kann vorher umgewandelt werden:


````python
eingabe = "9,99"
preis = float(eingabe.replace(",", "."))

print(preis)
````

## 11. Unterschied zwischen `=` und `==`

`=` weist einen Wert zu:

```python
alter = 18
```

`==` vergleicht zwei Werte:

```python
alter == 18
```

Das Ergebnis eines Vergleichs ist `True` oder `False`.


````python
alter = 18

print(alter == 18)
print(alter == 20)
````

> **Merksatz:** Ein `=` speichert. Zwei `==` vergleichen.


## 12. Arithmetische Operatoren

Python unterstützt diese Rechenoperatoren:

- `+` Addition
- `-` Subtraktion
- `*` Multiplikation
- `/` Division
- `//` Ganzzahldivision
- `%` Rest einer Division
- `**` Potenz


````python
zahl_1 = 10
zahl_2 = 5

ergebnis = zahl_1 + zahl_2
print("Addition:", ergebnis)

ergebnis = zahl_1 - zahl_2
print("Subtraktion:", ergebnis)

print("Multiplikation:", zahl_1 * zahl_2)
print("Potenz:", zahl_1 ** zahl_2)

division = zahl_1 / zahl_2
print("Division:", division)
print("Datentyp der Division:", type(division))

modulo = 10 % 3
print("Modulo:", modulo)

ganzzahldivision = 10 // 3
print("Ganzzahldivision:", ganzzahldivision)
print("Datentyp:", type(ganzzahldivision))
````

### Punkt vor Strich und Klammern

Python rechnet nach den üblichen Rechenregeln:


````python
print(2 + 3 * 4)       # 14
print((2 + 3) * 4)     # 20
````

> **Merksatz:** Mit Klammern kann ich eindeutig festlegen, was zuerst berechnet werden soll.


## 13. Division, Ganzzahldivision und Modulo

### Normale Division `/`

```python
10 / 3
```

Ergebnis:

```text
3.3333333333333335
```

Das Ergebnis ist in der Regel ein `float`.

### Ganzzahldivision `//`

```python
10 // 3
```

Ergebnis:

```text
3
```

### Modulo `%`

Modulo liefert den Rest einer Division.

```python
10 % 3
```

Ergebnis:

```text
1
```


````python
print(10 / 3)
print(10 // 3)
print(10 % 3)
````

## 14. Zuweisungsoperatoren

Eine Variable kann mit ihrem bisherigen Wert weitergerechnet werden.


````python
zahl = 10

zahl += 5
print(zahl)

zahl -= 2
print(zahl)

zahl *= 3
print(zahl)

zahl /= 2
print(zahl)
````

Die Kurzformen bedeuten:

- `zahl += 1` → `zahl = zahl + 1`
- `zahl -= 1` → `zahl = zahl - 1`
- `zahl *= 2` → `zahl = zahl * 2`
- `zahl /= 2` → `zahl = zahl / 2`


## 15. Wahrheitswerte beim Rechnen

In Python verhält sich:

- `True` wie `1`
- `False` wie `0`


````python
bool_1 = True
bool_2 = True

print(bool_1 + bool_2)
print(True + False)
print(False + False)
````

Das ist technisch möglich, sollte aber nur verwendet werden, wenn der Zusammenhang verständlich bleibt.


## 16. Genauigkeit bei Kommazahlen

Kommazahlen werden im Computer binär gespeichert.

Einige Dezimalzahlen können dadurch nicht exakt dargestellt werden.

Das Originalbeispiel addiert mehrere Kommazahlen:


````python
wert = 0

wert += 0.12
wert += 2.1
wert += 3.1
wert += 3.1
wert += 3.1
wert += 0.1
wert += 0.1
wert += 0.1
wert += 0.1
wert += 0.1
wert += 0.1
wert += 0.1
wert += 0.1
wert += 0.1
wert += 0.1
wert += 0.1

print("Ungefilterter Wert:", wert)
print("Gerundet:", round(wert, 2))
````

### Kleiner Ausblick: Das geht später auch kürzer

Die vielen Zeilen mit `wert += 0.1` sehen ehrlich gesagt ziemlich stumpf aus.
Python kann solche Wiederholungen mit einer **Schleife** deutlich kürzer schreiben.

Später sieht derselbe Teil so aus:

```python
for _ in range(11):
    wert += 0.1
```

Du musst diesen Code hier noch nicht auseinandernehmen. `for`, `range()` und auch das `_` werden im eigenen Schleifen-Notes ganz in Ruhe erklärt.

> **Merksatz:** Hier geht es gerade um die Ungenauigkeit von Kommazahlen - nicht darum, nebenbei schon Schleifen zu lernen.


### Vorsicht bei Multiplikation und Division

Auch mathematisch einfache Rechnungen können intern kleine Rundungsabweichungen zeigen.


````python
print(49.0 * (1.0 / 49.0))
print((1 / 49) * 49)
````

Für normale Ausgaben reicht häufig:

```python
round(wert, 2)
```

Bei besonders genauen Geldberechnungen kann später `Decimal` sinnvoll sein.


````python
print(round(5.9))   # 6
````

## Aufgabe 1: Zwei Variablen tauschen

### Aufgabe

Gegeben sind:

```python
variable_a = "Hallo"
variable_b = "Welt"
```

Tausche die Inhalte.

### Dein Versuch


### Musterlösung


````python
variable_a = "Hallo"
variable_b = "Welt"

print("Vorher:", variable_a, variable_b)

variable_a, variable_b = variable_b, variable_a

print("Nachher:", variable_a, variable_b)
````

### Warum funktioniert das?

Python merkt sich zuerst beide Werte auf der rechten Seite. Danach werden sie in umgekehrter Reihenfolge links gespeichert.

Aus:

```text
variable_a = "Hallo"
variable_b = "Welt"
```

wird also:

```text
variable_a = "Welt"
variable_b = "Hallo"
```

Man könnte dafür auch eine dritte Hilfsvariable verwenden. Die Kurzform ist in Python aber sehr üblich.
