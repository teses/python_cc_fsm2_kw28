# Dict Comprehension

Ein Dict Comprehension in Python ist eine kompakte Möglichkeit,
Dictionaries (Wörterbücher) zu erstellen – ähnlich wie List Comprehensions für Listen.

Syntax:

`{key_expr: value_expr for element in iterable if condition}`

- key_expr → Ausdruck für den Schlüssel
- value_expr → Ausdruck für den Wert
- element → Platzhalter für jedes Element in der Iteration
- iterable → Sammlung, über die iteriert wird
- condition (optional) → Filterbedingung

```python
# Quadratzahlen als Dictionary
squares = {x: x**2 for x in range(6)}
print(squares)
```

```python
# Wörter in ihre Länge abbilden
words = ["Python", "ist", "cool", "ist"]
word_length = { word:len(word) for word in words}
print(word_length)
```

```python
# Filtern: Nur gerade Zahlen
even_squares = {x: x**2 for x in range(10) if x % 2 == 0}
print(even_squares)
```

```python
# Umkehren eines Dictionaries
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print(inverted)
```
