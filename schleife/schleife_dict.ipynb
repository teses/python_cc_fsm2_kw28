```python
person = {
    "firstName": "Susi",
    "lastName": "Sauer",
    "gender": "w",
    "height": 163,
    "weight": 60.6
}
```

alle keys durchlaufen

```python
for key in person.keys():
    print(key, person[key])
```

alle values durchlaufen. keine keys!

```python
for value in person.values():
    print(value)
```

alle durchlaufen mit key und value

```python
for key, value in person.items():
    print(f"{key} = {value}")
```

heute nimmt man enumerate()

```python
for key, value in enumerate(person):
    print(f"{key} = {value}")
```
