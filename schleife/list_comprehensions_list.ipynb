# List Comprehesions mit if

```python
# Filter gerade Zahlen
even_numbers = [x for x in range(1, 51) if x % 2 == 0]
print(even_numbers)
```

```python
zahlen = range(0, 21)
zahlen2 = [zahl for zahl in zahlen if zahl > 10]
print(list(zahlen))
print(zahlen2)
```

# Kombination Conditional Expressions mit List- Comprehensions

```python
x = 2
a = "gerade" if x % 2 == 0 else "ungerade"
print(a)
```

```python
labels = [
    "gerade" if x % 2 == 0 else "ungerade"
    for x in range(0, 11)
]
print(labels)
```

# Ein paar Beispiele

```python
# Beispiel: alle Wörter groß
woerter = ["python", "java", "c++"]
gross = [
    w.upper()
    for w in woerter
]
print(gross)
```

# Beispiel Verschachtelt
- Verschachtelte Schleifen
- mit vorsicht zu verwenden. wird schnell unleserlich
- wenn man schachtelt -> NICHT unbedingt in einer Zeile quetschen

```python
#paare = [(x, y) for x in range(0,3) for y in range(0,3)] # eine zeile

paare = [
    (x, y)
    for x in range(3)
    for y in range(3)
]
print(paare)
```

```python
# mit bedingungen
paare = [
    (x, y)
    for x in range(3) if x == 2
    for y in range(3) if y == 1
]
print(paare)
```
