Schleifen in Python

while -Kopfgesteuerte schleife

Fussgesteuerte Schleifen gibt es nicht


```python
i = 0
while i < 10:
    print(i)
    i += 1
```

while mit else

```python
i = 0
while i < 10:
    print(i)
    i += 1
else:
    print("i ist nicht mehr kleiner als 10")
```

Break-Schlüsselwort

```python
# Das break Schlüsselwort
i = 0
while i < 10:
    print(i)
    if i == 5:
        break
    i += 1
```

Fussgesteuerte Schleife

```python
i = 1
while(True):
    print(i)
    i += 1
    if i >= 10:
        break
```

Continue-Schlüsselwort

```python
i = 0
while i < 10:
    i += 1
    if i % 2 == 0:
        continue
    print(i)
```
