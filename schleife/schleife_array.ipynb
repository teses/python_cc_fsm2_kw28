# Wie läuft man elegant ein array durch

normal mit while

```python

obst = ["Apfel", "Banane", "Kiwi"]
anzahl = len(obst)
i = 0
while i < anzahl:
    print(obst[i])
    i += 1
```

mit for schleife => elegant

```python
obstkorb = ["Apfel", "Banane", "Kiwi"]
for obst in obstkorb:
    print(obst)
```

index manuel basteln

```python
obstkorb = ["Apfel", "Banane", "Kiwi"]
i=0
for obst in obstkorb:
    print(f"Nr. {i} = {obst}")
    i += 1
```

Alte art und weise wenn man ein index und den Wert haben will

```python
obstkorb = ["Apfel", "Banane", "Kiwi"]
for i in range(len(obstkorb)):
    print(f"Nr. {i} = {obstkorb[i]}")
```

mit enumerate() holt man index und wert auf eleganter art und weise. pythonic

```python
obstkorb = ["Apfel", "Banane", "Kiwi"]
for no, obst in enumerate(obstkorb, start=1):
    print(f"Nr. {no} = {obst}")
```

```python
obstkorb = ["Apfel", "Banane", "Kiwi"]
for i, obst in enumerate(obstkorb, start=0):
    print(f"index {i} = {obst}")
```
