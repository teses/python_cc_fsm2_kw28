List Comprehensions in Python sind eine kompakte und elegante Möglichkeit, Listen zu erstellen.
Statt mit einer Schleife Elemente einzeln zu einer Liste hinzuzufügen, können wir mit einer einzigen Zeile Listen generieren.

Syntax
    `[expression for element in iterable if condition]`

- expression → was in die neue Liste kommt
- element → Platzhalter für jedes Element in der Iteration
- iterable → Datenquelle (Liste, String, Range, etc.)
- if condition (optional) → Filter

```python
zahlen = range(1, 11)
liste = [zahl*2 for zahl in zahlen]
print(list(zahlen))
print(liste)
```

```python
# verdoppelung
liste = [zahl*2 for zahl in range(1, 11)]
print(liste)
```

Vergleich zwischen normaler schleife und Comprehesions

```python
liste = [x*x for x in range(0 , 10)]
print(liste)

#dies als normale schleife
liste=[]
for x in range(0 , 10):
    liste.append(x*x)
print(liste)
```
