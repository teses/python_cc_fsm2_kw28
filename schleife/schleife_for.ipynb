Normale for schleife

```python
autos = ["VW", "Audi", "Porsche"]
for auto in autos:
    print(auto)
```

```python
zahlen = range(2, 10, 2)
for zahl in zahlen:
    print(zahl)
```

```python
# Alle Buchstaben von A-Z
for x in range(65, 91):
    print(chr(x))
```

Verschachtelung von for schleifen

```python
farben = ["blauer", "gelber", "grüner"]
obst = ["Apfel", "Pfirsich", "Kirsche"]
for farbe in farben:
    for frucht in obst:
        print(farbe + " " + frucht)
```
