Dictionaries erlauben als Wert jeden anderen Datentyp =>

- Dictionary in dictionary
- Eine liste kann dictionaries enthalten
- Dictionarties können listen enthalten

```python
import json

def p_print(data):
    return print(json.dumps(data, indent=4))

mg = {
    "Hans": 1,
    "Jürgen": 0
}
kl = {
    "Ben": 2,
    "Tim": 6
}
teams = {
    "Mönchengladbach": mg,
    "Köln": kl
}

#############################
teams = {
    "Mönchen gladbach": {
        "Hans": 1,
        "Jürgen": 0
    },
    "Köln": {
        "Ben": 2,
        "Tim": 6
    }
}

p_print(teams)
print(teams["Mönchen gladbach"]["Hans"])

```
