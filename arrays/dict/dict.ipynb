# leserliche ausgabe mit JSON

```python
import json

def p_print(data):
    return print(json.dumps(data, indent=4))


person = {
    "firstName": "Susi",
    "lastName": "Sauer",
    "gender": "w",
    "height": 163,
    "weight": 60.6,
    "hobbies": ["singen", "tanzen"],
    "hair" :"blond"
}
print(person)
#print(json.dumps(person, indent=4))
p_print(person)
```

```python
# zugriff - variante 1
print(person["firstName"])
print(person["weight"])
print(person["hobbies"][1])
print(person["firstName"] + " " + person["lastName"])
```

```python
# member ändern
person["lastName"] = "Müller"
p_print(person)
```

```python
# member löschen
person.pop("hair")
print(person)
```

```python
address = {
    "street": "Musterstr. 1",
    "zip": "1234",
    "city": "Entenhausen"
}
alles = {**person, **address}
p_print(alles)
```
