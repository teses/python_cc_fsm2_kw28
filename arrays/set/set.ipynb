# set

- kein index
- nicht geordnet
- nicht änderbar aber hinzufügbar und entfernbar (add/remove)
- keine duplikate

NICHT als zufallsgenerator nutzen!

```python
# Die Reihenfolge ist unbestimmt
meinSet = {"Apfel", "Zitrone", "Banane"}
print(meinSet)
```

```python
# Achtung der doppelte Eintrag verschwindet
meinSet2 = {"Apfel", "Zitrone", "Banane", "Zitrone", True, 1, 2}
print(meinSet2)
```

```python
# hinzufügen
meinSet.add("Melone")
print(meinSet)
```

```python
# mehrere mit update()
meinSet3 = {"Ananas", "Kiwi"}
meinSet.update(meinSet3)
print(meinSet)
```

```python
# entfernen mit remove(). Bei nicht existens gibt es ein Fehler
meinSet.remove("Pfirsich")
print(meinSet)
```

```python
# entfernen mit discard(). Bei nicht existens gibt es ein KEIN Fehler
meinSet.discard("safas")
print(meinSet)
```

```python
meinSet = {"Apfel", "Zitrone", "Banane", "Kiwi"}
meinSet3 = {"Ananas", "Kiwi"}
neuesSet = meinSet.union(meinSet3)
print(neuesSet)
```

```python
# Schnittmenge bilden mit intersection()
meinSet = {"Apfel", "Zitrone", "Banane", "Kiwi"}
meinSet3 = {"Apfel", "Ananas", "Kiwi"}
neuesSet = meinSet.intersection(meinSet3)
print(neuesSet)
```

```python
# Elemente die in A sind aber NICHT in B
meinSetA = {"Apfel", "Zitrone", "Banane", "Kiwi"}
meinSetB = {"Apfel", "Ananas", "Kiwi"}
neuesSet = meinSetA.difference(meinSetB)
print(neuesSet)
```

```python
# produkte
original = {"PE-123456", "PE-123457", "PE-123458", "PE-123459"}
neue = {"PE-123456", "PE-123457", "PE-123460"}

# etl - updateinfos
products_to_delete = original.difference(neue)
print("delete", products_to_delete)
products_to_update = original.intersection(neue)
print("update", products_to_update)
products_to_insert = neue.difference(original)
print("insert", products_to_insert)
```
