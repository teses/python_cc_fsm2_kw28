```python
#
meineListe = ("Apfel", "Banane", "Zitrone", "Birne", "Kiwi")
```

```python
# ein Wert anhängen Trick 1
tmp = list(meineListe)
tmp.append("Melone")
meineListe = tuple(tmp)
print(meineListe)
```

```python
# ein Wert anhängen Trick 2
newListe = ("Aprikose",)
meineListe = meineListe + newListe
print(meineListe)
```

# unpacking

```python
(var1, var2) = ("wert1", "wert2")

print(var1)
print(var2)
```

```python
abmessung = ("800", "600")
#
(width, height) = abmessung
print(width)
print(height)
```

```python
def macheRechteck():

    width = 800
    height = 600

    return (width, height)


(w, h) = macheRechteck() # direktes unpacking des returns
print(w)
print(h)
```
