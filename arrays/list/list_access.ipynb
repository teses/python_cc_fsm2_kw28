# list

In Python wird ein notmales Array "list" genannt

- Zugriff über Index,
- sortiert (ordered)
- veränderbar
- duplikate erlaubt

```python
Obstkorb = ["Apfel", "Banane", "Zitrone", "Birne", "Kiwi", "Banane"]
```

*Ausgabe mit Datentyp*

```python
print(Obstkorb, type(Obstkorb))
```

*Menge*

```python
print(len(Obstkorb))
```

# Slicing (Teile aus dem Array holen)

```python
# Zugriff auf Listenelemente
print(Obstkorb[1])
print(Obstkorb[0])

# Zugriff auf Listenelemente mit negativen index
print(Obstkorb[-2])

# slice zum ende
print(Obstkorb[2:])

# slice vom anfang
print(Obstkorb[:5])

# Zugriff auf Bereich
print(Obstkorb[3:5])

# negativ => range von hinten
print(Obstkorb[-3:])
print(Obstkorb[-5:-2])

# der dritte wert nach den : steht für die Schrittweite
print(Obstkorb[0:5:2])

# Array drehen
print(Obstkorb[::-1])
```
