# list Methoden

```python
Obstkorb = ["Banane", "Zitrone", "Birne", "Banane", "Kiwi", "Apfel", "zweschge", "affe"]
```

```python
# ein Wert anhängen
Obstkorb.append("Melone")
print(Obstkorb)
```

```python
# entfernen - remove - Achtung es wird nur immer der erste Wert entfernt!!!
Obstkorb.remove("Banane")
print(Obstkorb)
```

```python
# löschen anhand des Index
Obstkorb.pop(1)
Obstkorb.pop() # den letzten löschen
print(Obstkorb)
```

```python
# einfügen
Obstkorb.insert(2, "Kirsche")
print(Obstkorb)
```

```python
# Liste erweitern mit einer anderen Liste - zwei listen zusammenfügen
meineListe = ["VW", "Audi", "BMW"]
meineListe.extend(Obstkorb)
print(Obstkorb)
print(meineListe)
```

```python
# Man kann listen auch einfach addieren
meineListe2 = ["VW", "Audi", "BMW"]
neueListe2 = meineListe2 + Obstkorb
print(neueListe2)
```

in python gibt es einen sogenannten unpacking operator. `*` und `**`

z.B. bei functionen die (*args, **kwargs)

```python
meineListe3 = ["VW", "Audi", "BMW"]
neueListe3 = [*meineListe3, *Obstkorb]
print(neueListe3)
```

# Sortierung

```python
# sortieren
Obstkorb.sort(reverse=False) # sortiert case sensitiv
Obstkorb.sort(key=str.lower)
print(Obstkorb)
```

```python
Obstkorb.sort(key=len)
print(Obstkorb)
```

```python
def len_sort(obst):
    # 20 zeilen code
    return len(obst)

Obstkorb.sort(key=len_sort)
#Obstkorb.sort(key=lambda obst: len(obst))
print(Obstkorb)
```
