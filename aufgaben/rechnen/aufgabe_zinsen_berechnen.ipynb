# Zinsen berechnen

Schreibe ein Programm, das ein Startkapital, einen Zinssatz und die Anzahl der Jahre speichert.

Berechne das Endkapital nach Zinseszinsen.

`Endkapital = Startkapital * (( 1 +  Zinssatz/100 ) ** Jahre)`


```python
startkapital = float(input("Geben Sie das Startkapital in Euro ein: "))
zinssatz = float(input("Geben Sie den Zinssatz in % ein: "))
jahre = int(input("Geben Sie die Laufzeit in Jahren ein: "))

#algorithmus
endkapital = startkapital * (1 + zinssatz / 100) ** jahre

#gewinn
zinsgewinn = endkapital - startkapital


print("\n" + "=" * 40)
print("     ZINSESZINS-BERECHNUNG     ")
print("=" * 40)
print(f"Startkapital:   {startkapital:10.2f} €")
print(f"Zinssatz:       {zinssatz:10.2f} %")
print(f"Laufzeit:       {jahre:10d} Jahre")
print("-" * 40)
print(f"Zinsgewinn:     {zinsgewinn:10.2f} €")
print(f"Endkapital:     {endkapital:10.2f} €")
print("=" * 40)
```

```python
# Gegeben:
startkapital = 3000
zinssatz = 10
jahre = 1

# Ausgabe:
# Endkapital: 1157.625
endkapital = startkapital * ((1 + (zinssatz / 100)) ** jahre)

print(round(endkapital, 3))
```
