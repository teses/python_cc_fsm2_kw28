# Ein Programm zur Berechnung der Sozialversicherung

Gegeben sind folgende Werte
- Bruttolohn in Euro
- Arbeitslosenversicherung in %
- Rentenversicherung in %
- Krankenversicherung in %
- Pflegeversicherung in %
- (Version 2: Die Angabe der Anzahl der Kinder)

Als Ergebnis sollen folgende Werte angezeigt werden:
- Bruttolohn in Euro
- Betrag der für die Arbeitslosenversicherung abgezogen wird
- Betrag der für die Rentenversicherung abgezogen wird
- Betrag der für die Krankenversicherung abgezogen wird
- Betrag der für die Pflegeversicherung abgezogen wird
- Das zu versteuernde Einkommen

Die Beitragsbemessungsgrenze soll ert mal nicht betrachtet werden
Dies kann in einer weiteren Verion eingebaut werden

Die Kirchensteuer soll nicht betrachtet werden.
Dies kann in einer weiteren Verion eingebaut werden

```python
# Eingabe
bruttolohn = 3000

prozent_av = 1.3  # Arbeitslosenversicherung
prozent_rv = 9.3  # Rentenversicherung
prozent_kv = 7.3  # Krankenversicherung (inkl. Zusatzbeitrag)
prozent_pv = 2.3  # Pflegeversicherung Grundbeitrag

anzahl_kinder = 0

# Verarbeitung
if anzahl_kinder == 0:
    prozent_pv += 0.6


betrag_av = bruttolohn * (prozent_av / 100)
betrag_rv = bruttolohn * (prozent_rv / 100)
betrag_kv = bruttolohn * (prozent_kv / 100)
betrag_pv = bruttolohn * (prozent_pv / 100)

gesamt_abzuege = betrag_av + betrag_rv + betrag_kv + betrag_pv

zu_versteuerndes_einkommen = bruttolohn - gesamt_abzuege

# Ausgabe
print("\n" + "=" * 50)
print("ERGEBNIS DER SOZIALVERSICHERUNGSBERECHNUNG")
print("=" * 50)
print(f"Bruttolohn:                               {bruttolohn:10.2f} €")
print("-" * 50)
print(f"Arbeitslosenversicherung ({prozent_av:.1f}%):       {betrag_av:10.2f} €")
print(f"Rentenversicherung ({prozent_rv:.1f}%):             {betrag_rv:10.2f} €")
print(f"Krankenversicherung ({prozent_kv:.1f}%):            {betrag_kv:10.2f} €")
print(f"Pflegeversicherung ({prozent_pv:.1f}%):             {betrag_pv:10.2f} €")
print("-" * 50)
print(f"Gesamte Sozialabzüge:                     {gesamt_abzuege:10.2f} €")
print(f"Zu versteuerndes Einkommen:               {zu_versteuerndes_einkommen:10.2f} €")
print("=" * 50)
```

```python
# 2. Lösung
# Zuweisung Variablen
bruttolohn = 3000
arbeitslosenversicherung_prozent = 1.3
rentenversicherung_prozent = 9.3
krankenversicherung_prozent = 7.3
pflegeversicherung_prozent = 2.9

# Berechnung
arbeitslosenversicherung_euro = round(bruttolohn * (arbeitslosenversicherung_prozent / 100), 2)
rentenversicherung_euro = round(bruttolohn * (rentenversicherung_prozent / 100), 2)
krankenversicherung_euro = round(bruttolohn * (krankenversicherung_prozent / 100), 2)
pflegeversicherung_euro = round(bruttolohn * (pflegeversicherung_prozent / 100), 2)
zu_versteuerndes_einkommen = round(bruttolohn - arbeitslosenversicherung_euro - rentenversicherung_euro - krankenversicherung_euro - pflegeversicherung_euro, 2)


# Ausgabe
print("Bruttolohn in Euro: ", bruttolohn)
print("- Arbeitslosenversicherung in Euro: ", arbeitslosenversicherung_euro)
print("- Rentenversicherung in Euro: ", rentenversicherung_euro)
print("- Krankenversicherung in Euro: ", krankenversicherung_euro)
print("- Pflegeversicherung in Euro: ", pflegeversicherung_euro)
print("Das zu versteuernde Einkommen in Euro: ", zu_versteuerndes_einkommen)
```
