# Programmieraufgabe (Rabatte & Zusatzkosten beim Online-Einkauf)

1. Der Warenwert (in €) ist gegeben.
2. Vom Warenwert werden folgende Rabatte und Zuschläge berechnet:
    - Mengenrabatt: 10 %
    - Sonderrabatt: 5 %
    - Versandkosten: 4,90 € (fester Betrag)
    - Verpackungskosten: 2 %
3. Auf den neuen Betrag nach Abzügen und Zuschlägen kommt zusätzlich die Mehrwertsteuer von 19 % drauf.

Ausgabe:

Das Programm soll am Ende übersichtlich ausgeben:
- den eingegebenen Warenwert,
- den abgezogenen Mengenrabatt,
- den abgezogenen Sonderrabatt,
- die hinzugerechneten Versandkosten,
- die hinzugerechneten Verpackungskosten,
- die berechnete Mehrwertsteuer,
- den Endpreis nach allen Abzügen und Zuschlägen.

```python
#Wert
warenwert = 100.99

#Rabatte und Zuschläge berechnen
mengenrabatt = warenwert * 0.10  # 10 %
sonderrabatt = warenwert * 0.05  # 5 %
versandkosten = 4.90  # fester Betrag in €
verpackungskosten = warenwert * 0.02  # 2 %

#Zwischensumme berechnen
zwischensumme = (
    warenwert - mengenrabatt - sonderrabatt + versandkosten + verpackungskosten
)

#Mwst
mwst = zwischensumme * 0.19

endpreis = zwischensumme + mwst

print("\n" + "=" * 45)
print("     ÜBERSICHT IHRES ONLINE-EINKAUFS     ")
print("=" * 45)
print(f"Warenwert:                 {warenwert:8.2f} €")
print(f"- Mengenrabatt (10 %):     {mengenrabatt:8.2f} €")
print(f"- Sonderrabatt (5 %):      {sonderrabatt:8.2f} €")
print(f"+ Versandkosten:           {versandkosten:8.2f} €")
print(f"+ Verpackungskosten (2 %): {verpackungskosten:8.2f} €")
print("-" * 45)
print(f"Zwischensumme (Netto):     {zwischensumme:8.2f} €")
print(f"+ Mehrwertsteuer (19 %):   {mwst:8.2f} €")
print("=" * 45)
print(f"ENDPREIS (Brutto):         {endpreis:8.2f} €")
print("=" * 45)
```

```python
# Eingabe
warenwert = 100.99
mengenrabattProzent = 10
sonderrabattProzent = 5
versandkostenEuro = 4.90
verpackungskostenProzent = 2
mwstProzent = 19

# prozent in euro umrechnen
mengenrabatt = warenwert * (mengenrabattProzent/100)
sonderrabatt = warenwert * (sonderrabattProzent/100)
verpackungskosten = warenwert * (verpackungskostenProzent/100)

# eurowerte abziehen/aufschlagen
zwischensumme = warenwert - mengenrabatt - sonderrabatt + versandkostenEuro + verpackungskosten

# mwst berechnen
mehrwertsteuer =  zwischensumme*mwstProzent/100

# endpreis
endpreis = zwischensumme + mehrwertsteuer

# Ausgabe
print("\n--- Preisberechnung ---")
print(f"Warenwert:            {warenwert:8.2f} €")
print(f"- Mengenrabatt:       {mengenrabatt:8.2f} €")
print(f"- Sonderrabatt:       {sonderrabatt:8.2f} €")
print(f"+ Versandkosten:      {versandkostenEuro:8.2f} €")
print(f"+ Verpackungskosten:  {verpackungskosten:8.2f} €")
print(f"= Zwischensumme:      {zwischensumme:8.2f} €")
print(f"+ Mehrwertsteuer:     {mehrwertsteuer:8.2f} €")
print(f"Endpreis:             {endpreis:8.2f} €")
```
