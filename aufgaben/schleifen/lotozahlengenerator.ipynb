# Lottozahlen Generator

Es soll ein Lottozahlen Generator gebaut werden für die Zahlen
6 aus 49

6 Zahlen sollen gezogen werden. Diese dürfen nicht doppelt vorkommen

Die Lösung soll in einer schleife sein.

Die sample() Methode soll nicht genutzt werden

```python
import random

lottozahlen = []

while len(lottozahlen) < 6:
    gezogene_zahl = random.randint(1, 49) # Zufallszahl zwischen 1 und 49
    if gezogene_zahl not in lottozahlen:
        lottozahlen.append(gezogene_zahl)

lottozahlen.sort()

print("Gezogene Lottozahlen:", lottozahlen)
```

# Lösung 2 als Funktion

```python
import random

def lottozahlen():
    gezogene = []  # Liste für die gezogenen Zahlen
    while len(gezogene) < 6:
        zahl = random.randint(1, 49)  # Zufallszahl zwischen 1 und 49
        if zahl not in gezogene:      # doppelte Zahlen vermeiden
            gezogene.append(zahl)
    gezogene.sort()
    return gezogene

# Beispielaufruf
print("Ihre Lottozahlen sind:", lottozahlen())
```
