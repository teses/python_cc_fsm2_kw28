Stelle dir vor, du hast eine Klasse mit 4 Reihen, und jede Reihe hat 5 Sitzplätze.

Die Plätze sollen in einer verschachtelten Liste gespeichert werden. Jeder Platz enthält den Namen eines Schülers (als String).

1. Erstelle eine verschachtelte Liste sitzplan, in der die Namen der Schüler gespeichert sind.
2. Schreibe ein Programm, das:
	- Alle Namen in Reihe 2 (Index 1) ausgibt.
	- Den 3. Sitzplatz in Reihe 3 (Index 2, Sitz 2) ausgibt.
	- Alle Schüler Spalte 1 (also den ersten Sitz jeder Reihe) ausgibt.
	- Einen bestimmten Schüler durch "Frei" ersetzt (z. B. "Nina" → "Frei").

```python
sitzplan = [
    ["Anna", "Ben", "Cem", "Diana", "Elif"],
    ["Felix", "Greta", "Hassan", "Isabel", "Jonas"],
    ["Kaan", "Lena", "Mert", "Nina", "Oskar"],
    ["Paul", "Quentin", "Rita", "Sofia", "Tom"]
]

# 1. Alle Namen aus Reihe 2
print("Reihe 2:", sitzplan[1])

# 2. Dritter Sitzplatz aus Reihe 3
print("Dritter Sitzplatz in Reihe 3:", sitzplan[2][2])

# 3. Alle Schüler auf Platz 1 jeder Reihe
print("1. Sitzplatz jeder Reihe:")
for reihe in sitzplan:
    print(reihe[0])


# 4. Schüler "Nina" durch "Frei" ersetzen
for reihe in sitzplan:
    for platz_index, name in enumerate(reihe):
        if name == "Anna":
            reihe[platz_index] = "Frei"

print("\nSitzplan nach der Änderung:")
for reihennummer, reihe in enumerate(sitzplan, start=1):
    print(f"Reihe {reihennummer}: {reihe}")

```

weitere Aufgaben zu der Schülerliste
- Finde alle Schüler, deren Name mit "A" beginnt.
- Zähle, wie viele Schüler insgesamt im Sitzplan sind.
- Erstelle eine flache Liste mit allen Namen.
- Suche einen bestimmten Schüler (z. B. "Mert") und finde sein Reihen-Index und Platz-Index, als auch die Reihe und den Platz als Nr.

```python
sitzplan = [
    ["Anna", "Albert", "Cem", "Diana", "Elif"],
    ["Felix", "Greta", "Hassan", "Isabel", "Jonas"],
    ["Kaan", "Lena", "Mert", "Nina", "Oskar"],
    ["Paul", "Quentin", "Rita", "Sofia", "Tom"]
]

# Finde alle Schüler, deren Name mit "A" beginnt.
namen_mit_a = []
for reihe in sitzplan:
    for platz in reihe:
        if platz.startswith("A"):
            namen_mit_a.append(platz)

print(namen_mit_a)
```

```python
#Zähle, wie viele Schüler insgesamt im Sitzplan sind.
anzahl = 0
for reihe in sitzplan:
    anzahl += len(reihe)

print("Anzahl Schüler:", anzahl)
```

```python
# Aufgabe 3: Erstelle eine flache Liste mit allen Namen.
alle_schueler = []
for reihe in sitzplan:
    alle_schueler.extend(reihe)

print(alle_schueler)
```

```python
alle_namen = []

for reihe in sitzplan:
    for name in reihe:
        alle_namen.append(name)

print("Flache Liste:", alle_namen)
```

```python
# Aufgabe 4: Suche einen bestimmten Schüler (z. B. "Mert") und finde sein Reihen-Index und Platz-Index, als auch die Reihe und den Platz als Nr.
sitzplan = [
    ["Anna", "Albert", "Cem", "Diana", "Elif"],
    ["Felix", "Greta", "Hassan", "Isabel", "Jonas"],
    ["Kaan", "Lena", "Mert", "Nina", "Oskar"],
    ["Paul", "Quentin", "Rita", "Sofia", "Tom"]
]
gesucht = "Mert"
gefunden = None

for reihe_i, reihe in enumerate(sitzplan):
    for spalte_i, platz in enumerate(reihe):
        if platz == gesucht:
            gefunden = (reihe_i, spalte_i, platz, f"Reihe {reihe_i+1}, Platz {spalte_i+1}")

print(gefunden)
```
