**Zeichenhäufigkeit. Tipp: Ein String kann man wie ein Array durchlaufen!**

````python
text = "Mississippi"
erg = { char : text.count(char) for char in text }
print(erg)
````

**Häufigkeit zählen mit Dict Comprehension**

````python
obst = ["Apfel", "Banane", "Apfel", "Kirsche", "Banane", "Apfel"]

haeufigkeit = {
    sorte: obst.count(sorte)
    for sorte in set(obst)
}

print(haeufigkeit)
````

**ASCII-Tabelle**

`chr()` wandelt Zahl in Buchstabe, `ord()` gibt ASCII-Code.

Ausgabe:
```
{'A': 65, 'B': 66, ..., 'Z': 90}
```

````python
ascii_tabelle = {
    chr(code): code
    for code in range(ord("A"), ord("Z") + 1)
}
print(ascii_tabelle)
````

````python
ascii_dict = {
    chr(i): i
    for i in range(128)
    if chr(i).isalpha() and chr(i) == chr(i).upper()
}
print(ascii_dict)
````

**Preise inkl. Mehrwertsteuer**

````python
prices_netto = {"Apfel": 10, "Banane": 0.50, "Kirsche": 0.10}
prices_vat = { k: round(v * 1.19, 2) for k,v in prices_netto.items()}
print(prices_netto)
print(prices_vat)
````
