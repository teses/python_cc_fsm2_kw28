Geben Sie mithilfe von Schleifen folgende Figuren aus Sternen auf dem Bildschirm aus.

Hinweis: Pro Schreibvorgang darf jeweils immer nur ein Stern ausgegeben werden.
```
*******	                    7 Sterne
 ***** 	                    5 Sterne
  ***  	                    3 Sterne
   *   	                    1 Sterne
```
Abschwächung:
Hinweis: Es muss nicht nur ein Schreibvorgang je Sternchen sein.

````python
anzahl_zeilen = 4

for zeile in range(anzahl_zeilen):
    leerzeichen = zeile
    sterne = 7 - 2 * zeile

    print(" " * leerzeichen + "*" * sterne)
````

````python
zeilen = [1, 3, 5, 7, 9, 1]
max_breite = 9

for anzahl_x in zeilen:

    leerzeichen = int((max_breite - anzahl_x) / 2)

    for _ in range(leerzeichen):
        print(" ", end="")

    for _ in range(anzahl_x):
        print("X", end="")

    print()
````

# Lösung 1

````python
hoehe = 5
gesamtbreite = 2 * hoehe - 1

for zeile in range(hoehe):
    anzahl_x = 2 * zeile + 1 # erzeugt die ungeraden Zahlen 1, 3, 5, 7 und 9.
    leerzeichen = (gesamtbreite - anzahl_x) // 2 # berechnet die notwendigen Leerzeichen für eine mittige Ausrichtung.

    print(" " * leerzeichen + "X" * anzahl_x)

# Stamm mittig unter der Figur
print(" " * (gesamtbreite // 2) + "X")
````
