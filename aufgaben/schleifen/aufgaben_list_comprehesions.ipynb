# Sitzplatzanalyse

Bei den normalen Schleifenaufgaben sind noch einige Aufgabenstellungen, die man mit list-Comprehesions umsetzen kann

```python
sitzplan = [
    ["Anna", "Albert", "Cem", "Diana", "Elif"],
    ["Felix", "Greta", "Hassan", "Isabel", "Jonas"],
    ["Kaan", "Lena", "Mert", "Nina", "Oskar"],
    ["Paul", "Quentin", "Rita", "Sofia", "Tom"]
]

# Summe aller plätze
#menge = ???

# Die Summe der Plätze in einer Reihe

# Finde alle Schüler, deren Name mit "A" beginnt.
namen_mit_a = [
    platz
    for reihe in sitzplan
    for platz in reihe
    if platz.startswith("A")
]
print(namen_mit_a)
```
