Erzeuge eine neue Liste:

- Werte ab 50 sollen "bestanden" ergeben
- Werte unter 50 sollen ""nicht bestanden" ergeben

```python
punkte = [35, 50, 72, 49]
ergebnisse = [
    "bestanden" if punktzahl >= 50 else "nicht bestanden"
    for punktzahl in punkte
]

print("Punkte:    ", punkte)
print("Ergebnisse:", ergebnisse)
```

Eine verschachtelte Liste in lineare Liste wandeln (flatten)
Was bedeuten flatten?

```python
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
flache_liste = [
    zahl
    for zeile in matrix
    for zahl in zeile
]

print(flache_liste)
```

**Wörter nach Länge filtern**

Erzeuge eine Liste aller Wörter, die länger als 3 Buchstaben haben, und wandle sie in Großbuchstaben um.

```python
words = ["house", "dog", "international", "cup", "AI"]
lange_woerter = [
    wort.upper()
    for wort in words
    if len(wort) > 3
]

print(lange_woerter)
```

Erstellt eine Liste mit n Zufallszahlen und sortiert sie.
Achtung Modul notwendig

```python
import random

n = 10

zufallszahlen = [
    random.randint(1, 1000)
    for _ in range(n)
]

sortierte_zahlen = sorted(zufallszahlen)

print("Unsortiert:", list(zufallszahlen))
print("Sortiert:  ", sortierte_zahlen)
```

# Sitzplatzanalyse

Bei den normalen Schleifenaufgaben sind noch einige Aufgabenstellungen, die man mit list-Comprehesions umsetzen kann

```python
sitzplan = [
    ["Anna", "Albert", "Cem", "Diana", "Elif"],
    ["Felix", "Greta", "Hassan", "Isabel", "Jonas"],
    ["Kaan", "Lena", "Mert", "Nina", "Oskar"],
    ["Paul", "Quentin", "Rita", "Sofia", "Tom"]
]

# Summe aller plätze
gesamtMenge = sum(len(i) for i in sitzplan)
print(gesamtMenge)

# Die Summe der Plätze in einer Reihe
mengeJeZeile = list(len(i) for i in sitzplan)
print(mengeJeZeile)

# Finde alle Schüler, deren Name mit "A" beginnt.
namen_mit_a = [
    platz
    for reihe in sitzplan
    for platz in reihe
    if platz.startswith("A")
]
print(namen_mit_a)
```

```python
sitzplan_lc = [
    ["Jason", "Barry", "Mark", "Peter", "Andreas", "Maik"],
    ["John", "Jane", "Susi", "Sara", "Sven"],
    ["Max", "Moritz", "Mia", "Anna", "Markus"],
    ["Lukas", "Lena", "Joshua", "Luke", "Oliver"],
]
# Anzahl der Plätze je Reihe
plaetze_je_reihe = [len(reihe) for reihe in sitzplan_lc]
print(plaetze_je_reihe)

# Flache Liste aller Namen
alle_personen = [
    name
    for reihe in sitzplan_lc
    for name in reihe
]
print(alle_personen)

# Namen, die mit M beginnen
namen_mit_m = [
    name
    for reihe in sitzplan_lc
    for name in reihe
    if name.startswith("M")
]
print(namen_mit_m)
```
