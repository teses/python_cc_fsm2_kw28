```python
def status_anzeigen(gesamtplaetze, belegte_plaetze, zufahrt_blockiert, fuer_veranstaltung_angemietet):
    """Ermittelt den Status eines Parkhauses anhand der Entscheidungstabelle."""

    # Eingaben überprüfen
    if gesamtplaetze <= 0:
        print("Fehler: Die Anzahl der Gesamtplätze muss größer als 0 sein.")
        return

    if belegte_plaetze < 0 or belegte_plaetze > gesamtplaetze:
        print("Fehler: Die belegten Plätze liegen außerhalb des gültigen Bereichs.")
        return

    auslastung = belegte_plaetze / gesamtplaetze * 100

    # Sperrung hat laut Entscheidungstabelle Vorrang.
    if zufahrt_blockiert or fuer_veranstaltung_angemietet:
        status = "Parkhaus gesperrt"
    elif auslastung > 95:
        status = "Parkhaus belegt"
    else:
        status = "Parkhaus frei"

    print(f"Belegte Stellplätze: {belegte_plaetze} von {gesamtplaetze}")
    print(f"Auslastung: {auslastung:.1f} %")
    print(f"Status: {status}")


# Beispiel aus der Aufgabenstellung
testfaelle = [
    (500, 475, False, False),  # genau 95 %
    (500, 476, False, False),  # über 95 %
    (500, 100, True, False),   # blockiert
    (500, 100, False, True),   # angemietet
    (500, 500, False, True),   # voll und angemietet
]

for testfall in testfaelle:
    print("\n--- Testfall", testfall, "---")
    status_anzeigen(*testfall)
```

```python
def status_anzeigen(anzahl_plaetze_belegt, eingang_blockiert, ausgang_blockiert, vermietet):

    max_plaetze = 500 * 0.95

    if anzahl_plaetze_belegt > max_plaetze and not (eingang_blockiert or ausgang_blockiert) and not vermietet:
        print("Parkhaus belegt")
    elif anzahl_plaetze_belegt <= max_plaetze and not (eingang_blockiert or ausgang_blockiert) and not vermietet:
        print("Parkhaus frei")
    else:
        print("Parkhaus gesperrt")




Auslastung = [496, 400, 550]
EinfahrtBlockiert = [True, False]
AusfahrtBlockiert = [True, False]
EventAngemietet = [True, False]

i=0
for fall1 in Auslastung:
    for fall2 in EinfahrtBlockiert:
        for fall3 in AusfahrtBlockiert:
            for fall4 in EventAngemietet:
                print(i, fall1, fall2, fall3, fall4, end= " --> ")
                status_anzeigen(fall1, fall2, fall3, fall4)
                i += 1
```
