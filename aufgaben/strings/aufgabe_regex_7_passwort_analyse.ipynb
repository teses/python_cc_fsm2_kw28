# Passwort analysieren

Prüfe, ob ein Passwort

- mindestens 8 Zeichen lang ist,
- mindestens einen Großbuchstaben enthält,
- mindestens einen Kleinbuchstaben enthält,
- mindestens eine Zahl enthält.

```python
import re

passwoerter = [
    "Sicheres123",    # ok (11 Zeichen, Groß, Klein, Zahl)
    "pass1234",       # falsch (kein Großbuchstabe)
    "PASS1234",       # falsch (kein Kleinbuchstabe)
    "SicheresPW",     # falsch (keine Zahl)
    "Sich1",          # falsch (zu kurz, nur 5 Zeichen)
    "Sicheres 123 ",  # ok, aber Vorsicht: enthält Leerzeichen (Regex erlaubt das hier)
]

# ^               -> Start
# (?=.*[A-Z])     -> Mindestens ein Großbuchstabe
# (?=.*[a-z])     -> Mindestens ein Kleinbuchstabe
# (?=.*\d)        -> Mindestens eine Zahl
# .{8,}           -> Mindestens 8 Zeichen
# $               -> Ende
pattern = re.compile(r"^(?=.*[A-Z])(?=.*[a-z])(?=.*\d).{8,}$")

#Testlauf
for pw in passwoerter:
    if pattern.match(pw):
        print(f"Gültig: {pw!r}")
    else:
        print(f"Ungültig: {pw!r}")
```
