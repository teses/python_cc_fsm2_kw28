# Erkennung eines deutschen KFZ Kennzeichens

# Kennzeichenliste zum testen

```python
kennzeichen = [
    "WĂś AB 12",  # falsch
    "B M 1",  # true
    "B MW 1",  # true
    "WES A 10",  # true
    "WES A 1000",  # true
    "WES A 1000E",  # true
    "DU ME 123" ,  # true
    "DU ME 123H",   # true
    " DU ME 123H ",   # falsch
    " DU ME 12-3H ",  # falsch
    " ABC A 123",      # falsch
    "WES A 0",  # true
]
```

# LĂ¶sung 1

```python
import re

# Das Regex-Pattern erklĂ¤rt:
# ^               -> Start
# [A-Z]{1,3}      -> 1 bis 3 GroĂźbuchstaben
# \s              -> ein Leerzeichen
# [A-Z]{1,2}      -> 1 bis 2 GroĂźbuchstaben
# \s              -> ein Leerzeichen
# [1-9][0-9]{0,3} -> Eine Zahl
# [EH]?           -> 'E' oder ein 'H'
# $               -> Ende
pattern = re.compile(r"^[A-ZĂ–Ă„Ăś]{1,3}\s[A-Z]{1,2}\s[1-9][0-9]{0,3}[EH]?$")

# Testlauf
for kz in kennzeichen:
    if pattern.match(kz):
        print(f"True:  {kz!r}")
    else:
        print(f"False: {kz!r}")
```
