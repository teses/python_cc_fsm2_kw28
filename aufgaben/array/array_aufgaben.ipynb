Aufgabe: Palindrom erkennen

```python
def ist_palindrom(text):
    # Alles in Kleinbuchstaben
    text = text.lower()

    # Prüfung
    return text == text[::-1]


# Testläuf
test_woerter = ["Lagerregal", "Otto", "Anna", "Python", "Rentner"]

for wort in test_woerter:
    if ist_palindrom(wort):
        print(f" '{wort}' ist ein Palindrom!")
    else:
        print(f" '{wort}' ist KEIN Palindrom.")


```

```python
def isPalindrome(w):
    return w.upper().replace(" ", "") == w[::-1].upper().replace(" ", "")


test_woerter = ["Lagerregal", "Otto", "Anna", "Maoam","Python", "Rentner", "Erika feuert nur untreue Fakire"]

for wort in test_woerter:
    if isPalindrome(wort):
        print(f" '{wort}' ist ein Palindrom!")
    else:
        print(f" '{wort}' ist KEIN Palindrom.")
```
