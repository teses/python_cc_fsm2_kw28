```python
cd_liste = [
    {
        "album": "Bad",
        "interpret": "Michael Jackson",
        "jahr": 1987,
        "tracks": [
            "Bad",
            "The Way You Make Me Feel",
            "Man in the Mirror"
        ]
    },
    {
        "album": "Bravo Hits 1",
        "interpret": "Various Artists",
        "jahr": 1992,
        "tracks": [
            "Rhythm Is a Dancer",
            "Please Don't Go",
            "It's My Life"
        ]
    },
    {
        "album": "Eigene Mischung",
        "interpret": "Verschiedene",
        "jahr": 2026,
        "tracks": ["Titel 1", "Titel 2", "Titel 3"]
    }
]

for cd in cd_liste:
    print(f'\n{cd["album"]} – {cd["interpret"]}')
    for nummer, track in enumerate(cd["tracks"], start=1):
        print(f"{nummer}. {track}")

```
