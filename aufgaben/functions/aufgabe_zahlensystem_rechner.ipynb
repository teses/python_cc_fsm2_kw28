# Umrechnerfunktion

- HEX Zahl soll als String behandelt werden
- Dezimalzahl soll als INTEGER behandelt werden
- Dualzahl soll als String behandelt werden

1. Dual zu Dezimal        Additionsverfahren
2. Dezimal zu Dual        Divisionsverfahren
3. Hex zu Dezimal         Additionsverfahren
4. Dezimal zu Hex         Divisionsverfahren
5. Dual zu Hex            dual -> dezimal -> hex
6. Hex zu Dual            hex -> dezimal -> dual


#3  = 3  = 0011
#9  = 9  = 1001
#A  = 10 = 1010

Nicht die eingebauten Funktionen benutzen.

Es sollen eigene Funktionen geschaffen werden.
