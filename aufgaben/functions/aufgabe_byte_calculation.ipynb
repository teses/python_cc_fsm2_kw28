Schreibe eine Funktion die bytes in kB, MB, GB oder TB ausgibt.
Die Ausgabe soll in einem menschenlesbaren Format dargestellt werden.
Die Ausgabe soll die Einheiten automatisch bestimmen.

z.B.
```
900             => 900 Bytes
1000            => 1 kB
1024            => 1,024 kB
10000           => 10 kB
100000          => 100 kB
1000000         => 1 MB
1000001         => 1,000 MB
1239000         => 1,239 MB
1239900         => 1,240 MB
1234400         => 1,234 MB
1234567         => 1,235 MB
1234567891      => 1,235 GB
1234567891234   => 1,235 TB
```

Regeln
- 3 stellen nach dem Komma

Aufgabe ist:
- Funktion schreiben
- Datentypen definieren
- Kommentare schreiben (Docstrings)
- Doctests schreiben

````python
import doctest


def byteCalculationDezimal(bytes: int) -> str:
    """Konvertiert eine Byte-Anzahl im Dezimalsystem (Faktor 1000) in eine

    menschenlesbare Zeichenkette (Bytes, kB, MB, GB, TB).

    >>> byteCalculationDezimal(900)
    '900 Bytes'
    >>> byteCalculationDezimal(1000)
    '1 kB'
    >>> byteCalculationDezimal(1024)
    '1,024 kB'
    >>> byteCalculationDezimal(1000000)
    '1 MB'
    >>> byteCalculationDezimal(1239000)
    '1,239 MB'
    >>> byteCalculationDezimal(1239900)
    '1,240 MB'
    >>> byteCalculationDezimal(1234567891)
    '1,235 GB'
    >>> byteCalculationDezimal(1234567891234)
    '1,235 TB'
    """
    einheiten = ["Bytes", "kB", "MB", "GB", "TB"]
    wert = float(bytes)

    for einheit in einheiten:
        if wert < 1000 or einheit == einheiten[-1]:
            # Wenn der Wert eine ganze Zahl ist (z.B. bei 1000 -> 1 kB), ohne Nachkommastellen anzeigen
            if wert.is_integer():
                return f"{int(wert)} {einheit}"
            else:
                # 3 Nachkommastellen erzwingen und Punkt durch Komma ersetzen
                formatted = f"{wert:.3f}".replace(".", ",")
                return f"{formatted} {einheit}"

        wert /= 1000.0


print(byteCalculationDezimal(1000))
print(byteCalculationDezimal(1000000))
print(byteCalculationDezimal(1000001))
print(byteCalculationDezimal(1239000))
print(byteCalculationDezimal(1234567891))
print(byteCalculationDezimal(1234567891234))
````

````python
def byteCalculationDezimal(zahl: int) -> str:

    if zahl < 1000:
        return f"{zahl} Bytes"

    if zahl < 1000000:
        wert = zahl / 1000
        einheit = "kB"

    elif zahl < 1000000000:
        wert = zahl / 1000000
        einheit = "MB"

    elif zahl < 1000000000000:
        wert = zahl / 1000000000
        einheit = "GB"

    else:
        wert = zahl / 1000000000000
        einheit = "TB"


    gerundet = round(wert, 3)


    if gerundet.is_integer():
        formatiert = str(int(gerundet))
    else:
        formatiert = f"{gerundet:.3f}".replace('.', ',')


    return f"{formatiert} {einheit}"



byte_werte = [
    900, 1000, 1024, 10000, 100000, 1000000,
    1000001, 1239000, 1239900, 1234400, 1234567,
    1234567891, 1234567891234
]

for werte in byte_werte:
    print(f"{werte} => {byteCalculationDezimal(werte)}")


print(byteCalculationDezimal(byte_werte[3]))
````

````python
def bytes_lesbar(anzahl_bytes: int) -> str:
    """
    Wandelt Bytes automatisch in eine passende SI-Einheit um.

    :param anzahl_bytes: Anzahl der umzurechnenden Bytes
    :return: Menschenlesbare Darstellung mit passender Einheit
    :raises TypeError: Wenn kein Integer übergeben wird
    :raises ValueError: Wenn die Anzahl negativ ist

    >>> bytes_lesbar(900)
    '900 Bytes'

    >>> bytes_lesbar(1000)
    '1,000 kB'
    >>> bytes_lesbar(1024)
    '1,024 kB'
    >>> bytes_lesbar(10_000)
    '10,000 kB'
    >>> bytes_lesbar(100_000)
    '100,000 kB'
    >>> bytes_lesbar(1_000_000)
    '1,000 MB'
    >>> bytes_lesbar(1_234_567)
    '1,235 MB'
    >>> bytes_lesbar(1_234_567_891)
    '1,235 GB'
    >>> bytes_lesbar(1_234_567_891_234)
    '1,235 TB'
    """

    if isinstance(anzahl_bytes, bool) or not isinstance(anzahl_bytes, int):
        raise TypeError("Die Anzahl der Bytes muss eine ganze Zahl sein.")

    if anzahl_bytes < 0:
        raise ValueError("Die Anzahl der Bytes darf nicht negativ sein.")

    if anzahl_bytes >= 1_000_000_000_000:
        wert = anzahl_bytes / 1_000_000_000_000
        einheit = "TB"

    elif anzahl_bytes >= 1_000_000_000:
        wert = anzahl_bytes / 1_000_000_000
        einheit = "GB"

    elif anzahl_bytes >= 1_000_000:
        wert = anzahl_bytes / 1_000_000
        einheit = "MB"

    elif anzahl_bytes >= 1_000:
        wert = anzahl_bytes / 1_000
        einheit = "kB"

    else:
        return f"{anzahl_bytes} Bytes"

    text = f"{wert:.3f}".replace(".", ",")

    return f"{text} {einheit}"

testwerte = [
    900,
    1000,
    1024,
    10000,
    100000,
    1000000,
    1000001,
    1239000,
    1239900,
    1234400,
    1234567,
    1234567891,
    1234567891234,
]

for wert in testwerte:
    print(f"{wert:>15} => {bytes_lesbar(wert)}")
````
